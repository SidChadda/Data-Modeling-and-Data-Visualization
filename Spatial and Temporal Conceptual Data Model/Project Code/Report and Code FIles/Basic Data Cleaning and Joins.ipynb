# Combine DataFrames

```python
Days_Combined = pd.DataFrame()
Key_Metar_Combined=pd.DataFrame()
weather_Metar_Combined=pd.DataFrame()
Days_Combined = Days_Combined.append([Day_18, Day_19, Day_20, Day_21, Day_22, Day_23,Day_24])

# # Add Row_ID
# Days_Combined['Row_ID'] = range(0, Days_Combined.shape[0])
# Days_Combined.loc[:, 'Row_ID'] = 'P_'+Days_Combined['Row_ID'].astype(str)



Key_Metar_Combined = Key_Metar_Combined.append([Day_18_Metar, Day_19_Metar, Day_20_Metar, Day_21_Metar, Day_22_Metar, Day_23_Metar,Day_24_Metar])



```

```python
# Drop Metar from Days combined
#reset index 
Days_Combined = Days_Combined.reset_index(drop=True)
Key_Metar_Combined = Key_Metar_Combined.reset_index(drop=True)

# Joining
Days_Combined_Joined=Days_Combined.join(Key_Metar_Combined)
Days_Combined_Joined=Days_Combined_Joined.drop(columns='Metar')
Days_Combined_Joined.head()
```

```python
# Break Tem and Dew point
Days_Combined_Joined[['Temperature','Dew Point']]=Days_Combined_Joined['Temp and Dew Point'].str.split("/",expand=True,)
Days_Combined_Joined=Days_Combined_Joined.drop(columns=['Temp and Dew Point'])

# Add Row_ID
Days_Combined_Joined['Row_ID'] = range(0, Days_Combined_Joined.shape[0])
Days_Combined_Joined.loc[:, 'Row_ID'] = 'P_'+Days_Combined_Joined['Row_ID'].astype(str)

# Remove 'M ' from visibility
Days_Combined_Joined['Visibility'] = Days_Combined_Joined['Visibility'].map(lambda x: x.strip('M'))


Days_Combined_Joined.head()
```
